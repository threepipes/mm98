```python
exp = df[['K', 'meanDist', 'noCapturedCO', 'deadKnightCO']]
exp['dens'] = df.M / (df.S ** 2)
obj = df.score / df.score_nochange - 1
```

```python
# sklearn.linear_model.LinearRegression クラスを読み込み
from sklearn import linear_model
clf = linear_model.LinearRegression()
 
X = exp.as_matrix()
Y = obj.as_matrix()
 
# 予測モデルを作成
clf.fit(X, Y)
 
# 回帰係数
print(clf.coef_)
 
# 切片 (誤差)
print(clf.intercept_)
 
# 決定係数
print(clf.score(X, Y))
```

```python
regr = (exp * clf.coef_).sum(axis=1) + clf.intercept_
result = pd.concat([regr, obj], axis=1)
result
```
